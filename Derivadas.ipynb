Derivada hacia adelante

$$
f' _i = \frac{f_{i+1}-f{_i}}{x_{i+1}-x_i}
$$

Derivada hacia adelante

$$
f' _i = \frac{f_{i+1}-f{_i}}{x_{i+1}-x_i}
$$

```python
def deriv_h2(x,f):
    # Esto funciona si la malla es uniforme
    f_prima = zeros_like(f)
    h = x[1]-x[0]
    for i in range(len(f)-1):
        f_prima[i] = (f[i+1]-f[i])
    f_prima[-1] = f[-1]-f[-2]
    return f_prima/h
```

```python
x = linspace(0,3*pi,50)
y = x*sin(x)
# derivada analitica
dy = sin(x)+x*cos(x)
dy_h2 = deriv_h2(x,y)
```

```python
figure(dpi=100)
plot(x,dy,label="f'")
plot(x,dy_h2,label="f' num√©rica")
legend()
show()
```

```python
figure(dpi=100)
plot(x,abs(dy-dy_h2),label="Error")
legend()
show()
```

Derivada centrada

$$
f' _i = \frac{f_{i+1}-f{_{i-1}}}{x_{i+1}-x_{i-1}}
$$

Derivada centrada

$$
f' _i = \frac{f_{i+1}-f{_{i-1}}}{x_{i+1}-x_{i-1}}
$$

```python
def deriv_h3_mix(x,f):
    # Ojo esta deivada en las orillas tiene menos precision
    f_prima = zeros_like(f)
    dosh , h = x[2]-x[0] , x[1]-x[0] 
    for i in range(1,len(f)-1):
        f_prima[i] = (f[i+1]-f[i-1])/dosh
    f_prima[0] = (f[1]-f[0])/h
    f_prima[-1] = (f[-1]-f[-2])/h
    return f_prima
```

```python
dy_h3m = deriv_h3_mix(x,y)
```

```python
figure(dpi=100)
plot(x,dy,label="f'")
plot(x,dy_h2,'--',label=r"f' num√©rica $h^2$")
plot(x,dy_h3m,label=r"f' num√©rica $h^3$ (frontera $h^2$)")
legend()
show()
```

```python
figure(dpi=100)
plot(x,abs(dy-dy_h2),'--',label=r"Error de f' num√©rica $h^2$")
plot(x,abs(dy-dy_h3m),label=r"Error de f' num√©rica $h^3$ (frontera $h^2$)")
legend()
show()
```

```python
def deriv_h3(x,f):
    # Ojo esta deivada en las orillas tiene menos precision
    f_prima = zeros_like(f)
    dosh , h = x[2]-x[0] , x[1]-x[0] 
    for i in range(1,len(f)-1):
        f_prima[i] = (f[i+1]-f[i-1])/dosh
    f_prima[0] = (-0.5*f[2] + 2*f[1] - 1.5*f[0])/h
    f_prima[-1] = -(-0.5*f[-3] + 2*f[-2] - 1.5*f[-1])/h
    return f_prima
```

```python
dy_h3 = deriv_h3(x,y)
```

```python
figure(dpi=100)
plot(x,dy,label="f'")
plot(x,dy_h2,'--',label=r"f' num√©rica $h^2$")
plot(x,dy_h3m,label=r"f' num√©rica $h^3$ (frontera $h^2$)")
plot(x,dy_h3,label=r"f' num√©rica $h^3$")
legend()
show()
```

```python
figure(dpi=100)
plot(x,abs(dy-dy_h2),'--',label=r"Error de f' num√©rica $h^2$")
plot(x,abs(dy-dy_h3m),label=r"Error de f' num√©rica $h^3$ (frontera $h^2$)")
plot(x,abs(dy-dy_h3),label=r"Error de f' num√©rica $h^3$")
legend()
show()
```
