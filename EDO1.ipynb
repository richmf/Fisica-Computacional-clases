```python
def mi_Euler(ec_EDO , t , x0 , delta=0.1):
    # ec_EDO es el sistema a integrar (t,x)
    # t es el intervalo de integracion
    # x0 punto inicial
    tiempos = arange(t[0],t[1]+delta,delta)
    N = len(tiempos)
    sol = zeros(N)
    sol[0] = x0
    for i in range(N-1):
        sol[i+1] = sol[i] + delta*ec_EDO(tiempos[i],sol[i])
    return sol,tiempos
```

$$
\frac{dy}{dt} = \cos(t)
$$
con $y(0)=\pi/2$. La solución es: $y(t)=\sin(t)+\pi/2$

```python
def ec_dif1(t,y):
    return cos(t)
```

```python
sol,t = mi_Euler(ec_dif1,[0,pi],pi/2,delta=0.1)
plot(t,sol,label='$\delta = 0.1$')
sol,t = mi_Euler(ec_dif1,[0,pi],pi/2,delta=0.01)
plot(t,sol,label='$\delta = 0.01$')
plot(t,sin(t)+pi/2,label='Sol. analítica')
legend()
show()
```

$$
\frac{dy^2}{dt^2} = -y
$$
con $y(0)=7$ y $y'(0) = 0$. La solución es: $y(t)=7\cos(t)$.

Hcemos $u = \frac{dy}{dt}$ y por lo tanto $\frac{du}{dt} = \frac{dy^2}{dt^2} = -y$. Ton's:

$$
\frac{dy}{dt} = u \\
\frac{du}{dt} = -y
$$
con condición inicial $(y,u) = (7,0)$

```python
def mi_Euler(ec_EDO , t , x0 , delta=0.1):
    # ec_EDO es el sistema a integrar (t,x)
    # t es el intervalo de integracion
    # x0 punto inicial
    tiempos = arange(t[0],t[1]+delta,delta)
    M , N = len(x0) , len(tiempos)
    sol = zeros( (M,N) )
    sol[:,0] = x0
    for i in range(N-1):
        sol[:,i+1] = sol[:,i] + delta*ec_EDO(tiempos[i],sol[:,i])
    return sol,tiempos
```

```python
def ec_dif3(t,y):
    return array([y[1],-y[0]])
```

```python
sol,t = mi_Euler(ec_dif3,[0,4*pi],[7,0],delta=0.1)
plot(t,sol[0],label='$\delta = 0.1$')
sol,t = mi_Euler(ec_dif3,[0,4*pi],[7,0],delta=0.01)
plot(t,sol[0],label='$\delta = 0.01$')
plot(t,7*cos(t),label='Sol. analítica')
legend()
show()
```

```python
sol,t = mi_Euler(ec_dif3,[0,4*pi],[7,0],delta=0.1)
plot(sol[0],sol[1],label='$\delta = 0.1$')
sol,t = mi_Euler(ec_dif3,[0,4*pi],[7,0],delta=0.01)
plot(sol[0],sol[1],label='$\delta = 0.01$')
plot(7*cos(t),-7*sin(t),label='Sol. analítica')
title('Espacio Fase')
legend()
show()
```

$$
\frac{dy^2}{dt^2} = -y + \cos(\omega t)
$$
con $y(0)=7$ y $y'(0) = 0$. 
Hacemos $u = \frac{dy}{dt}$ y por lo tanto $\frac{du}{dt} = \frac{dy^2}{dt^2} = -y + \cos(\omega t)$. Ton's:

$$
\frac{dy}{dt} = u \\
\frac{du}{dt} = -y + \cos(\omega t) 
$$
con condición inicial $(y,u) = (7,0)$

```python
def ec_dif4(t,y,omega=1.0):
    return array([y[1],-y[0]+cos(omega*t)])
```

```python
sol,t = mi_Euler(ec_dif4,[0,300*pi],[7,0],delta=0.001)
plot(t,sol[0],label='$\delta = 0.001$')
legend()
show()
```
