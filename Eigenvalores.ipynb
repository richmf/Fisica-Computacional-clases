```python
from pylab import *
```

```python
def Mat_A(x):
    N = len(x) - 2
    A = zeros((N,N))
    for i in range(N):
        A[i,i] = -2
    for i in range(N-1):
        A[i,i+1] = 1
        A[i+1,i] = 1
    return A
```

```python
x = linspace(0,1,50)
h = x[1]-x[0]
```

```python
A = -Mat_A(x)/h**2
e_val,e_vec = eig(A)
```

```python
sort(e_val) - (arange(1,49)*pi)**2
```

```python
e_val
```

```python
fact = 1/abs(e_vec[:,30]).max()
```

```python
plot(x,sin(pi*x),label='n=1 Analítica')
plot(x[1:-1],-fact*e_vec[:,30],'x',label='n=1 Numérica')
legend()
show()
```

```python
plot(x,sin(2*pi*x),label='n=2 Analítica')
plot(x[1:-1],fact*e_vec[:,31],'x',label='n=2 Numérica')
legend()
show()
```

```python
plot(x,sin(3*pi*x),label='n=3 Analítica')
plot(x[1:-1],-fact*e_vec[:,32],'x',label='n=3 Numérica')
legend()
show()
```
