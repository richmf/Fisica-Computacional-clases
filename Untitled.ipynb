```python
from pylab import *
```

```python
x = linspace(-2,2)
y = linspace(-2,2)
XX , YY = meshgrid(x,y)
z = exp(-(XX**2+YY**2))
```

```python
graf = figure()
ejes = graf.add_subplot(projection='3d')
ejes.plot_surface(XX,YY,z)
show()
```

```python
niveles = linspace(0,1)
contour(XX,YY,z,levels=niveles)
gca().set_aspect('equal')
show()
```

```python
contourf(XX,YY,z)
gca().set_aspect('equal')
show()
```

```python
x = linspace(-2,2)
y = linspace(-2,2)
XX , YY = meshgrid(x,y)
z = exp(-((XX-0.7)**2+YY**2)) + exp(-((XX+1)**2+YY**2))
```

```python
niveles = linspace(0,1.5,20)
contour(XX,YY,z,levels=niveles)
gca().set_aspect('equal')
show()
```

```python
graf = figure()
ejes = graf.add_subplot(projection='3d')
ejes.plot_surface(XX,YY,z)
show()
```
